```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
```

```python
# Mean: Average of values
# To find mean manually
population = [50,24,37,49,59,26]
sample = [6,7,3,5,2,4]

p_mean = sum(population)/len(population)
print(p_mean)

s_mean = sum(sample)/len(sample)
print(s_mean)
```

```python
# To find mean using numpy
data = [50,24,37,49,59,26]
a_mean = np.mean(data)
print(a_mean)
```

```python
# To find mean using statistics
import statistics as stat
a = [15,14,13,12,17,16]
stat_mean = stat.mean(a)
print(stat_mean)
```

```python
# To find mean on Real Dataset
iris = sns.load_dataset("iris")

mean_of_sepall = np.mean(iris["sepal_length"])
print(mean_of_sepall )
print(iris.head())
print(iris.shape)
```

```python
# Median: Middle value
data = [22,22,55,12,199,12,19]
med = np.median(data)
print(med)
```

```python
# Mode: Most repeated value
data = [10,20,20,20,20,10,10,14,6,7,3,2,410,10]
mode = stat.multimode(data)
print(mode)
```

```python
# Using scipy
from scipy import stats
data = [10,20,20,10,80,10,46,64]

mode = stats.mode(data)
print(mode)
```

```python
# Variance: How far data from mean
data = [2,5,6,6,7,5,9,14,10,2,3,4]
df = [2,4,6]

# ddof: Delta Degree of Freedom (If we are not apply this bydefault sample varience is calculated)
sample_variance = np.var(data,ddof=1)
population_variance = np.var(data,ddof=0) 

sample_variance1 = np.var(df,ddof=1) 
population_variance1 = np.var(df,ddof=0)

print(sample_variance)
print(population_variance)
print(sample_variance1)
print(population_variance1)
```

```python
# Through series
data = [2,5,6,6,7,5,9,14,10,2,3,4]

series = pd.Series(data)
sample_variance = series.var()
print(sample_variance)

population_variance = series.var(ddof=0)
print(population_variance)
```

```python
# Through stats
data = [2,5,6,6,7,5,9,14,10,2,3,4]

sample_variance = stats.tvar(data)
print(sample_variance)

sample_variance = stats.tvar(data,ddof=0)
print(sample_variance)
```

```python
# Standard Deviation
data = [2,5,6,6,7,5,9,14,10,2,3,4]

sample_std = np.std(data,ddof = 1)
print(sample_std)

population_std = np.std(data,ddof=0)
print(population_std)
```

```python
# Using pandas
data = [2,5,6,6,7,5,9,14,10,2,3,4]

series = pd.Series(data)
sample_std = series.std(ddof=1)
print(sample_std)

population_std = series.std(ddof=0)
print(population_std)
```

```python
# Using scipy
import math
data = [2,5,6,6,7,5,9,14,10,2,3,4]

sample_std = math.sqrt(stats.tvar(data,ddof=1))
print(sample_std)

population_variance = math.sqrt(stats.tvar(data,ddof=0))
print(population_variance)
```
